```python
data_frame.groupby('diagnosis').mean()
data_frame.drop(columns='Unnamed: 32',inplace=True)
```

```python
x=data_frame.drop('diagnosis',axis=1)
```

```python
y=data_frame['diagnosis']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
model=LogisticRegression()
```

```python
model
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
x_train_predict=model.predict(x_train)
training_prediction=accuracy_score(y_train,x_train_predict)
```

```python
training_prediction
```

```python
x_test_predict=model.predict(x_test)
test_prediction=accuracy_score(y_test,x_test_predict)
```

```python
#label encdoder refer to converting Labels into numeric form which may be in either numeric or string form as to convert it 
#into machine readable format
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df['Species']=le.fit_transform(df['Species'])#convert the string into numeric catoegory like speices cateogory into 0,1,2
df.head()
```

```python
from sklearn.model_selection import train_test_split
#train-70
#test-30
X=df.drop(columns=['Species'])
Y=df['Species']
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.30)
```

```python
#logistic regerssion
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
#model training
model.fit(x_train,y_train)
```

```python
#print metric to get performance
print("Accuracy",model.score(x_test,y_test))
```

```python
#knn-  k nearest neighbours
from sklearn.neighbors import  KNeighborsClassifier
model=KNeighborsClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
print("Acuracy =",model.score(x_test,y_test))
```

```python
print("Acuracy =",model.score(x_test,y_test)*100)
```

```python
#decision tree
from sklearn.tree import  DecisionTreeClassifier
model=DecisionTreeClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
print("Acuracy =",model.score(x_test,y_test))
```

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('sol.csv')
df
```

```python
y=df['logS']
y
```

```python
x=df.drop('logS',axis=1)
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=100)
```

```python
x_test #test has 20% data as defined by size=0.2
```

```python
x_train
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_lr_train_predict=lr.predict(x_train)
y_lr_test_predict=lr.predict(x_test)
```

```python
y_lr_train_predict
```

```python
y_lr_test_predict
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
lr_train_mse=mean_squared_error(y_train,y_lr_train_predict)
lr_train_r2=r2_score(y_train,y_lr_train_predict)

lr_test_mse=mean_squared_error(y_test,y_lr_test_predict)
lr_test_r2=r2_score(y_test,y_lr_test_predict)
```

```python
lr_train_mse
```

```python
lr_train_r2
```

```python
lr_test_mse
```

```python
lr_test_r2
```

```python
lr_results=pd.DataFrame(['Linear Regression',lr_train_mse,lr_train_r2,lr_test_mse,lr_test_r2]).transpose()
lr_results
```

```python
lr_results=pd.DataFrame(['Linear Regression',lr_train_mse,lr_train_r2,lr_test_mse,lr_test_r2]).transpose()
lr_results.columns=(['Methods','Training MSE','Training R2','Test MSE','Test R2'])
lr_results
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf=RandomForestRegressor(max_depth=2,random_state=100)
rf.fit(x_train,y_train)
```

```python
y_rf_train_predict=rf.predict((x_train))
y_rf_test_predict=rf.predict((x_test))
                              
```

```python
from sklearn.metrics import mean_squared_error,r2_score

rf_train_mse=mean_squared_error(y_train,y_rf_train_predict)
rf_train_r2=r2_score(y_train,y_rf_train_predict)

rf_test_mse=mean_squared_error(y_test,y_rf_test_predict)
rf_test_r2=r2_score(y_test,y_rf_test_predict)
```

```python
rf_results=pd.DataFrame(['Random Forest',rf_train_mse,rf_train_r2,rf_test_mse,rf_test_r2]).transpose()
rf_results.columns=(['Methods','Training MSE','Training R2','Test MSE','Test R2'])
rf_results
```

```python
df_model=pd.concat([lr_results,rf_results],axis=0)
df_model
```
